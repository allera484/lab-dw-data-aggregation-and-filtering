1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
import pandas as pd
pd.set_option('display.max_columns', 50)
df1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
df1.head(1)
```

```python
df1.columns
```

```python
#total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign
filter_claim = df1['Total Claim Amount']>1000
filter_marketing = df1['Response']=='Yes'
df1.loc[filter_claim&filter_marketing]
```

```python
df1.query('`Total Claim Amount` >1000 and Response == "Yes"')
```

```python
df1.groupby(df1['Total Claim Amount']>1000).get_group(True).groupby('Response').get_group('Yes')
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
pd.pivot_table(df1.loc[filter_marketing],index=['Policy Type','Gender'],values='Total Claim Amount',aggfunc='mean')
```

```python
df1.loc[filter_marketing].groupby(['Policy Type','Gender'])[['Total Claim Amount']].mean()
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
states_500 = df1['State'].value_counts()[df1['State'].value_counts()>500].index
filter_state = df1['State'].isin(states_500)
df1[filter_state].groupby('State').size()
df1[filter_state]['State'].value_counts()
pd.pivot_table(df1[filter_state],index='State',values='Customer',aggfunc='count')
```

```python
df1['State'].value_counts()[df1['State'].value_counts()>500]
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
from datetime import timedelta
timedelta()
```

```python
pd.Timedelta(53,unit='D')
```

```python
df1['Days Since Policy Inception'] = (df1['Months Since Policy Inception']*(365/12)).apply(timedelta)
```

```python
df1['Policy Creation'] = pd.to_datetime(df1['Effective To Date'])-df1['Days Since Policy Inception']
```

```python
df1['Month'] = df1['Policy Creation'].dt.month
```

```python
pd.pivot_table(df1,index='State',columns='Month',values='Number of Policies',aggfunc='sum')
```

```python
top_3_states = df1.groupby('State')['Number of Policies'].sum().sort_values(ascending=False).head(3).index
df1.loc[df1['State'].isin(top_3_states)].groupby(['State','Month'])['Number of Policies'].sum()
pd.pivot_table(df1.loc[df1['State'].isin(top_3_states)], index='State',columns='Month',values='Number of Policies',aggfunc='sum')
```
